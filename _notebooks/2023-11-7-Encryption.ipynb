  const binaryArray = binary.match(/.{1,8}/g);

    function textToBinary(text) {
        return text.split('').map(function (char) {
            return char.charCodeAt(0).toString(2).padStart(8, '0');
        }).join(' ');
    }
    function binaryRotate(binary, rotationAmount) {
        console.log(binary[binary.length-1]);
        // Example of binary rotation (move the last character to the beginning)
        var rotatedBinary=binary;
        for(let i=0;i<rotationAmount;i++){
            if(rotationAmount%2==0){
            rotatedBinary=rotateonce(rotatedBinary);
            }
            else{
                rotatedBinary=rotateForward(rotatedBinary);
            }
        }
        return rotatedBinary;
    }
    function rotateonce(binary){
        var newstring="";
        newstring+=binary[binary.length-1];
        for(let i=0;i<binary.length-1;i++){
            newstring+=binary[i];
        }
        return newstring;
    }
    function rotateForward(binary){
        var newstring = binary[binary.length - 1];
        for (let i = 1; i < binary.length; i++) {
            newstring += binary[i];
        }
        newstring+=binary[0];
        return newstring;
    }

        for (let i = 0; i < rotationAmount; i++) {
            rotatedBinary = rotateonceBackward(rotatedBinary);

    function speechtotext(){
        if ('SpeechRecognition' in window || 'webkitSpeechRecognition' in window) {
            const recognition = new (window.SpeechRecognition || window.webkitSpeechRecognition)();
            // Set continuous to true for continuous recognition
            recognition.continuous = true;
            // Set language (optional)
            recognition.lang = 'en-US';
            const startButton = document.getElementById('startButton');
            const outputDiv = document.getElementById('output');
            startButton.addEventListener('click', () => {
                recognition.start();
                startButton.disabled = true;
                startButton.textContent = 'Listening...';
            });
            recognition.onresult = (event) => {
                const result = event.results[event.resultIndex];
                const transcript = result[0].transcript;
                outputDiv.textContent += transcript + ' ';
            };
            recognition.onend = () => {
                startButton.disabled = false;
                startButton.textContent = 'Start Speech Recognition';
            };
        } else {
            alert('Speech recognition is not supported in this browser. Please use a supported browser.');
        }
