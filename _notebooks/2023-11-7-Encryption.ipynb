<script>
    function textToBinary(text) {
        return text.split('').map(function (char) {
            return char.charCodeAt(0).toString(2).padStart(8, '0');
        }).join(' ');
    }

    function binaryRotate(binary, rotationAmount) {
        rotationAmount = rotationAmount % 8; // Ensure rotation is within one byte
        return binary.substring(rotationAmount) + binary.substring(0, rotationAmount);
    }

    function encryptText() {
        // Get the input text and rotation amount
        var inputText = document.getElementById("inputText").value;
        var rotationAmount = parseInt(document.getElementById("rotationAmount").value);

        // Check if the rotation amount is a valid number
        if (isNaN(rotationAmount)) {
            alert("Please enter a valid rotation amount.");
            return;
        }

        // Convert text to binary
        var binaryText = textToBinary(inputText);

        // Rotate the binary representation
        var rotatedBinary = binaryRotate(binaryText, rotationAmount);

        // Display the encrypted text (binary rotation)
        document.getElementById("encryptedText").textContent = rotatedBinary;
    }
</script>
