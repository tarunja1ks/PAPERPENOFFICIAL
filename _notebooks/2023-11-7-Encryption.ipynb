    function textToBinary(text) {
        return text.split('').map(function (char) {
            return char.charCodeAt(0).toString(2).padStart(8, '0');
        }).join(' ');
    }

    function binaryRotate(binary, rotationAmount) {
        console.log(binary[binary.length-1]);
        
        

        // Example of binary rotation (move the last character to the beginning)
        var rotatedBinary=binary;
        for(let i=0;i<rotationAmount;i++){
           rotatedBinary=rotateonce(rotatedBinary);
        }
        return rotatedBinary;
    }

    function rotateonce(binary){
        var newstring="";
        newstring+=binary[binary.length-1];
        for(let i=0;i<binary.length-1;i++){
            newstring+=binary[i];
        }
        return newstring;

        var rotatedBinary = binaryRotate(binaryText, rotationAmount);
